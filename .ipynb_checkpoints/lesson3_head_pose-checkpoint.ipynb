## Regression with BIWI head pose dataset

This is a more advanced example to show how to create custom datasets and do regression with images. Our task is to find the center of the head in each image. The data comes from the [BIWI head pose dataset](https://data.vision.ee.ethz.ch/cvl/gfanelli/head_pose/head_forest.html#db), thanks to Gabriele Fanelli et al. We have converted the images to jpeg format, so you should download the converted dataset from [this link](https://s3.amazonaws.com/fast-ai-imagelocal/biwi_head_pose.tgz).

> ## 저희가 여기서 쓰는 코드는 `get_ctr`이고, 
> ## get_ctr은 input으로 filename을 받고, output으로는 tensor type을 줍니다.

```python
path = untar_data(URLs.BIWI_HEAD_POSE)
#download the data from the URL
```

```python
cal = np.genfromtxt(path/'01'/'rgb.cal', skip_footer=6); cal
```

```python
fname = '09/frame_00667_rgb.jpg'
#bring a file.
```

```python
def img2txt_name(f): return path/f'{str(f)[:-7]}pose.txt'
```

```python
img = open_image(path/fname)
img.show()
#show an image which we brought a few cells before.
```

```python
ctr = np.genfromtxt(img2txt_name(fname), skip_header=3); ctr
```

```python
def convert_biwi(coords):
    c1 = coords[0] * cal[0][0]/coords[2] + cal[0][2]
    c2 = coords[1] * cal[1][1]/coords[2] + cal[1][2]
    return tensor([c2,c1])

def get_ctr(f):
    ctr = np.genfromtxt(img2txt_name(f), skip_header=3)
    return convert_biwi(ctr)

def get_ip(img,pts): return ImagePoints(FlowField(img.size, pts), scale=True)
```

```python
get_ctr(fname)
```

```python
ctr = get_ctr(fname)
img.show(y=get_ip(img, ctr), figsize=(6, 6))
```

## Creating a dataset

```python
data = (PointsItemList.from_folder(path)
        .split_by_valid_func(lambda o: o.parent.name=='13')
        .label_from_func(get_ctr)
        .transform(get_transforms(), tfm_y=True, size=(120,160))
        .databunch().normalize(imagenet_stats)
       )
#create a new databunch
```

```python
data.show_batch(3, figsize=(9,6))
#show some data 
```

## Train model

```python
learn = cnn_learner(data, models.resnet34)
#bring the resnet 34 model
```

```python
learn.lr_find()
learn.recorder.plot()
#find the fastest model and check it
```

```python
lr = 2e-2
#learing rate = 2e-2
```

```python
learn.fit_one_cycle(3, slice(lr))
#train the model for 3 times
```

```python
learn.save('stage-1')
#save it
```

```python
learn.load('stage-1');
#bring it again
```

```python
learn.show_results()
#show the results
```

## Data augmentation

```python
tfms = get_transforms(max_rotate=20, max_zoom=1.5, max_lighting=0.5, max_warp=0.4, p_affine=1., p_lighting=1.)

data = (PointsItemList.from_folder(path)
        .split_by_valid_func(lambda o: o.parent.name=='13')
        .label_from_func(get_ctr)
        .transform(tfms, tfm_y=True, size=(120,160))
        .databunch().normalize(imagenet_stats)
       )
#new data bunch
```

```python
def _plot(i,j,ax):
    x,y = data.train_ds[0]
    x.show(ax, y=y)

plot_multi(_plot, 3, 3, figsize=(8,6))
```

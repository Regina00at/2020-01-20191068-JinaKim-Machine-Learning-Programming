## Looking at the data

We are going to use the [Oxford-IIIT Pet Dataset](http://www.robots.ox.ac.uk/~vgg/data/pets/) by [O. M. Parkhi et al., 2012](http://www.robots.ox.ac.uk/~vgg/publications/2012/parkhi12a/parkhi12a.pdf) which features 12 cat breeds and 25 dogs breeds. Our model will need to learn to differentiate between these 37 distinct categories. According to their paper, the best accuracy they could get in 2012 was 59.21%, using a complex model that was specific to pet detection, with separate "Image", "Head", and "Body" models for the pet photos. Let's see how accurate we can be using deep learning!

We are going to use the `untar_data` function to which we must pass a URL as an argument and which will download and extract the data.

```python
help(untar_data)
```

The **help ()** function shows a summary version of the document. It mainly tells us where the modules come from, what it does, and so on. In this case, **fname:Union[pathlib.Path, str]=None** means that file name is the type of path or string and defaults to nothing. Anyway, 'untar_data' downloads something from the path and untar it to return the value of the path.

```python
path = untar_data(URLs.PETS); path
```

So the **path** would be the fname of **URLs.PETS**. In python, writting the variable on its own prints the variable. So "path" here would be **/root/.fastai/data/oxford-iiit-pet**.

```python
path.ls()
```

**path.ls()** function shows what's in the path. The path is said to have annotations and images.

```python
path_anno = path/'annotations'
path_img = path/'images'
```

**/** is a very convenient function of Python 3 from pathlib. It lets you access the subpaths much easier.

The first thing we do when we approach a problem is to take a look at the data. We _always_ need to understand very well what the problem is and what the data looks like before we can figure out how to solve it. Taking a look at the data means understanding how the data directories are structured, what the labels are and what some sample images look like.

The main difference between the handling of image classification datasets is the way labels are stored. In this particular dataset, labels are stored in the filenames themselves. We will need to extract them to be able to classify the images into the correct categories. Fortunately, the fastai library has a handy function made exactly for this, `ImageDataBunch.from_name_re` gets the labels from the filenames using a [regular expression](https://docs.python.org/3.6/library/re.html).

```python
fnames = get_image_files(path_img)
fnames[:5]
#type(fnames)
```

**fnames = get_image_files(path_img)** : Get the image file paths from the image dataset. 

**fnames[:5]** : Since "fnames" is a list of file names, so we can index the paths in it. This codes instructs compiler to list the first five image paths.

```python
np.random.seed(2)
pat = r'/([^/]+)_\d+.jpg$'
```

**np.random.seed (2)** is the selection of random blocks of verification set data. Each time the code is reproduced, "ImageDataBunch" will automatically selects an arbitrary data chunk from the original data set.

**pat = r'/([^/]+)_\d+.jpg$'** : A regular expression  to be used to extract the label from the file name

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=224, bs=bs
                                  ).normalize(imagenet_stats)
```

**ImageDataBunch.from_name_re** : **DataBunch object** contains our training and validation dataset, optionally test dataset. In this case, imagedatabunch contains all of the data we need to build a model. 

**size** : GPU has to apply the same instruction for a bunch of things to be fast. Thus, images have to be in one same size and same shape. 224 by 224 is a very common size since final layer of data is 7 by 7.

**Transforms** : If not a 224 pixel image, the image should be cut and resized.

```python
data.show_batch(rows=3, figsize=(7,6))
```

**data.show_batch()** function shows the contents in the data bunch.

```python
print(data.classes)
len(data.classes),data.c
```

print() function shows what it is. Since **data.c** is the number of classes, the compiler shows the same result as the 'len(data.classes)'.

## Training: resnet34

Now we will start training our model. We will use a [convolutional neural network](http://cs231n.github.io/convolutional-networks/) backbone and a fully connected head with a single hidden layer as a classifier. Don't know what these things mean? Not to worry, we will dive deeper in the coming lessons. For the moment you need to know that we are building a model which will take images as input and will output the predicted probability for each of the categories (in this case, it will have 37 outputs).

We will train for 4 epochs (4 cycles through all our data).

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

Learner is training model in fastai. We put model or archituecture on the **models.resnet34** slot. **Resnet** is a pre-trained good model. The number behind it is the size of the model(how big it is). **Metrics** is about printing. 

```python
learn.model
```

**learn.model** : train model

```python
learn.fit_one_cycle(4)
```

**fit_one_cycle()** function fits the model going through the entire dataset. The number inside the function is how many times we go through the dataset. After 4 times of fitting, this model correctly pick the right one in 92% of the time.

```python
learn.save('stage-1')
```

**learn.save()** save the fitted model.

## Results

Let's see what results we have got. 

We will first see which were the categories that the model most confused with one another. We will try to see if what the model predicted was reasonable or not. In this case the mistakes look reasonable (none of the mistakes seems obviously naive). This is an indicator that our classifier is working correctly. 

Furthermore, when we plot the confusion matrix, we can see that the distribution is heavily skewed: the model makes the same mistakes over and over again but it rarely confuses other categories. This suggests that it just finds it difficult to distinguish some specific categories between each other; this is normal behaviour.

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()
#losses
#idxs
#len(data.valid_ds)==len(losses)==len(idxs)
```

**interp = ClassificationInterpretation.from_learner(learn)** : Interpret two information
1. what is the data
2. what is the model

**losses,idxs = interp.top_losses()** : figure out high/top losses to find out what's went wrong.


```python
interp.plot_top_losses(9, figsize=(15,11))
```

**interp.plot_top_losses** display top losses

```python
doc(interp.plot_top_losses)
```

**doc()** function shows the document of the function, method, and so on. The summary version of doc() is help() as we mentioned before.

**plot_confusion_matrix** draws a diagram to show what was wrong or right. It can easily display all the information about the matching.

## Unfreezing, fine-tuning, and learning rates

Since our model is working as we expect it to, we will *unfreeze* our model and train some more.

```python
learn.unfreeze()
```

**unfreeze()** is the thing that says please train the whole model. 

**[QUESTION2]** The statement above is what a detailed lesson note told me. What does "unfreeze()" exactly do?

> ## 좋은 질문입니다.
> ## deeplearning은 동영상에서 설명한 것과 같이 layer들이 층층이 쌓여 있어, 매우 단순한 부분부터 학습해, 이를 이용하여 더 고도의 것까지 학습하는 과정을 가집니다.
> ## 저희가 사용하는 fast.ai가 제공하는 모델들은 이미 기본적인 layer들을 학습을 완료해 놓은 모델들입니다.(이번 exercise에서는 cnn_learner)
> ## 이를 가져와 위에서 저희가 불러온 고양이 dataset을 학습하기 위해 fit_one_cycle(4)를 하는 거고요. 그리고 그 학습된 모델을 'stage-1'에 저장합니다.
> ## 하지만, 위 밑의 부분들은 저희가 새로이 처음부터 training(학습)을 시키는 부분이기에, 이미 학습된 내용들을 없애기 위해 unfreeze를 하는 것입니다.
> ## fit_one_cycle을 하는 이유는 다시 학습하기 위해서입니다!!

```python
learn.fit_one_cycle(1)
```

**fit_one_cycle()** : train the model. In this case, it train the model once again.

```python
learn.load('stage-1');
```

Since the error rate increased - which means the result got worse -, load previous model(just 4 times fitted one) using **load()** function.

```python
learn.lr_find()
```

**lr_find()** : **Learning rate finder** : This is the function that figures out what is the fastest model I can train.

```python
learn.recorder.plot()
```

**recorder.plot()** : This function shows the result of learning rate finder as a graph.

```python
learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(1e-6,1e-4))
```

Since **recorder.plot()** showed that the "loss" tends to be low in the range of 1e-6 to 1e-4, we'll train the model within this range. **Unfreeze()** the model and train it**(fit_one_cycle())** twice.

That's a pretty accurate model!

## Training: resnet50

Now we will train in the same way as before but with one caveat: instead of using resnet34 as our backbone we will use resnet50 (resnet34 is a 34 layer residual network while resnet50 has 50 layers. It will be explained later in the course and you can learn the details in the [resnet paper](https://arxiv.org/pdf/1512.03385.pdf)).

Basically, resnet50 usually performs better because it is a deeper network with more parameters. Let's see if we can achieve a higher performance here. To help it along, let's us use larger images too, since that way the network can see more detail. We reduce the batch size a bit since otherwise this larger network will require more GPU memory.

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(),
                                   size=299, bs=bs//2).normalize(imagenet_stats)
```

The method to train the model is the same. The difference is that the model has been changed into the larger one, resnet 50. So set the data again using **ImageDataBunch()**.

```python
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

**cnn_learner** : Download the "resnet50" model and data

```python
learn.lr_find()
learn.recorder.plot()
```

**lr_find()** : Let's see where the loss would be the lowest.

**recorder.plot()** : Display the result.

```python
learn.fit_one_cycle(8)
```

**fit_one_cycle()** : Train the model.

```python
learn.save('stage-1-50')
```

**save()** : Save.

It's astonishing that it's possible to recognize pet breeds so accurately! Let's see if full fine-tuning helps:

```python
learn.unfreeze()
learn.fit_one_cycle(3, max_lr=slice(1e-6,1e-4))
```

**unfreeze()** again and fit the model**(fit_one_cycle)** within the range of (1e-6,1e-4).

If it doesn't, you can always go back to your previous model.

```python
learn.load('stage-1-50');
```

**load()** : load model. In this case, load previous model.

```python
interp = ClassificationInterpretation.from_learner(learn)
```

**interp = ClassificationInterpretation.from_learner(learn)** : Interpret two information
1. what is the data
2. what is the model

```python
interp.most_confused(min_val=2)
```

**most_confused()** : Let's see what was the most confusing breed.

Using a regular expression, we can grab the label.
The first 2 lines **(pat = r"/(\d)/\d+\.png$";
data = ImageDataBunch.from_name_re(path, fn_paths, pat=pat, ds_tfms=tfms, size=24))** is for that. **classes()**; print.

```python
data = ImageDataBunch.from_name_func(path, fn_paths, ds_tfms=tfms, size=24,
        label_func = lambda x: '3' if '/3/' in str(x) else '7')
data.classes
```

Get data from name function; fn_paths; **ImageDataBunch.from_name_func()**.

**[QUESTION4]** What does "lambda"" exactly mean in the ImageDataBunch.from_name_func()?

> ## lambda 는 anonymous function으로서, 함수를 메모리에 저장할 필요 없이 생성하고 바로 사라지는 함수입니다.
> ## ImageDataBunch.from_name_func는 위의 다른 method들처럼 dataset에 labelling을 하는 함수입니다.
> ## 여기서 사용한 람다함수는, file name에 3이 있으면 label을 3, 아니면 7을 붙이는 역할을 합니다.

```python
labels = [('3' if '/3/' in str(x) else '7') for x in fn_paths]
labels[:5]
```

**[('3' if '/3/' in str(x) else '7') for x in fn_paths]** : List the label - 3 or 7.

**labels[:5]** : print the first five labels.

```python
data = ImageDataBunch.from_lists(path, fn_paths, labels=labels, ds_tfms=tfms, size=24)
data.classes
```

**ImageDataBunch.from_lists()** : SET THE DATA

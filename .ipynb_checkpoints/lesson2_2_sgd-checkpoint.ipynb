#  Linear Regression problem

The goal of linear regression is to fit a line to a set of points.

```python
n=100
```

```python
x = torch.ones(n,2) 
#torch.ones() : built-in function.
#ones(size, out=None, dtype=None, layout=torch.strided, device=None, requires_grad=False) -> Tensor
#size : a sequence of integers defining the shape of the output tensor.
#out : (Tensor, optional): the output tensor.
x[:,0].uniform_(-1.,1)
#uniform(low:numbers.Number, high:numbers.Number=None, size:Union[List[int], NoneType]=None) -> Union[float, torch.Tensor]
#Draw 1 or shape=`size` random floats from uniform dist: min=`low`, max=`high`.
x[:5]
```

```python
a = tensor(3.,2); a
#tensor(x:Any, *rest) -> torch.Tensor
#Like `torch.as_tensor`, but handle lists too, and can pass multiple vector elements directly.
```

```python
y = x@a + torch.rand(n)
#rand(*size, out=None, dtype=None, layout=torch.strided, device=None, requires_grad=False) -> Tensor
```

[QUESTION] what does '@' mean?

> ## 이 부분은 [행렬의 곱셈](https://zetawiki.com/wiki/행렬의_곱셈)을 알아야 합니다.
> ## x라는 100X2(100 rows and 2 columns)와 a(2 rows and 1 column)을 곱하는 연산자입니다.

```python
plt.scatter(x[:,0], y);
#scatter : draw a scatter type graph
```

You want to find **parameters** (weights) `a` such that you minimize the *error* between the points and the line `x@a`. Note that here `a` is unknown. For a regression problem the most common *error function* or *loss function* is the **mean squared error**. 

```python
def mse(y_hat, y): return ((y_hat-y)**2).mean()
#define a function
```

Suppose we believe `a = (-1.0,1.0)` then we can compute `y_hat` which is our *prediction* and then compute our error.

```python
a = tensor(-1.,1)
#tensor(x:Any, *rest) -> torch.Tensor
#Like `torch.as_tensor`, but handle lists too, and can pass multiple vector elements directly.
```

```python
y_hat = x@a
mse(y_hat, y)
#return ((y_hat-y)**2).mean()
```

```python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_hat);
#draw two scatter type graphs
```

So far we have specified the *model* (linear regression) and the *evaluation criteria* (or *loss function*). Now we need to handle *optimization*; that is, how do we find the best values for `a`? How do we find the best *fitting* linear regression.

# Gradient Descent

We would like to find the values of `a` that minimize `mse_loss`.

**Gradient descent** is an algorithm that minimizes functions. Given a function defined by a set of parameters, gradient descent starts with an initial set of parameter values and iteratively moves toward a set of parameter values that minimize the function. This iterative minimization is achieved by taking steps in the negative direction of the function gradient.

Here is gradient descent implemented in [PyTorch](http://pytorch.org/).

```python
a = nn.Parameter(a); a
```

```python
help(nn)
```

[QUESTION] I cannot find that what 'nn' is and why 'nn' is needed.

> ## 이 코드는 사실 반환하는 tensor의 값은 [-1,1]의 행렬로 같습니다만, gradient를 적용하기 위해 수행해준 코드입니다.

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
#define the update()
```

```python
lr = 1e-1
for t in range(100): update()
#100 times of update() function
```

```python
plt.scatter(x[:,0],y)
#draw scatter type of graph
with torch.no_grad():
    plt.scatter(x[:,0],x@a);
    #draw scatter type of graph
```
